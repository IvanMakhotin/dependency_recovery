```python
def norm2(X):
    return np.sum(X*X)

def kernel(X, X0, a):
    return math.e ** (-a*norm2(X-X0))

def Q(X, sample, lmbd):
    return sum([lmbd[i]*kernel(X,sample[i], 1) for i in range(len(sample))])
```

```python
def learn(sample1, sample2):
    lmbd1 = np.ones(sample1.shape[0])
    lmbd2 = np.ones(sample2.shape[0])
    errors = sample1.shape[0]
    while(errors > 0):
        errors = 0
        #print lmbd1, lmbd2
        for i in range(sample1.shape[0]):
            print "Q1 =", Q(sample1[i], sample1, lmbd1), " Q2 =", Q(sample1[i], sample2, lmbd2)
            if Q(sample1[i], sample1, lmbd1) <= Q(sample1[i], sample2, lmbd2):
                lmbd1[i] += 1
                lmbd2[i] -= 1
                errors += 1
        print " "
        for i in range(sample2.shape[0]):
            print "Q1 =", Q(sample2[i], sample1, lmbd1), " Q2 =", Q(sample2[i], sample2, lmbd2)
            if Q(sample2[i], sample1, lmbd1) > Q(sample2[i], sample2, lmbd2):
                lmbd1[i] -= 1
                lmbd2[i] += 1
                errors +=1
        print errors
    return (lmbd1, lmbd2)

def exam(sample1, sample2, exam1, exam2, lmbd1, lmbd2):
    errors = 0
    for i in range(exam1.shape[0]):
        print "Q1 =", Q(exam1[i], sample1, lmbd1), " Q2 =", Q(exam1[i], sample2, lmbd2)
        if Q(exam1[i], sample1, lmbd1) <= Q(exam1[i], sample2, lmbd2):
            errors += 1
    print " "
    for i in range(exam2.shape[0]):
        print "Q1 =", Q(exam2[i], sample1, lmbd1), " Q2 =", Q(exam2[i], sample2, lmbd2)
        if Q(exam2[i], sample1, lmbd1) > Q(exam2[i], sample2, lmbd2):
            errors +=1
    print errors
```
